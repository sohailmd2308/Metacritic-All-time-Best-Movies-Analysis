```python
# Importing the the required libraries

import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup
import ssl
import requests
import re

# Ignore SSL certificate errors
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
```

```python
# Below are the pages we iterate to get the movie names and movie links

#https://www.metacritic.com/browse/movies/score/metascore/all/filtered?sort=desc&page=0
#https://www.metacritic.com/browse/movies/score/metascore/all/filtered?sort=desc&page=1
#https://www.metacritic.com/browse/movies/score/metascore/all/filtered?sort=desc&page=2
#https://www.metacritic.com/browse/movies/score/metascore/all/filtered?sort=desc&page=3
#https://www.metacritic.com/browse/movies/score/metascore/all/filtered?sort=desc&page=4
```

```python
#100 movies are present in each page
#page 0 to 4 i.e. 500 movies
# use a list all_urls to append all the urls 

all_urls = []
for i in range(5):
    url = 'https://www.metacritic.com/browse/movies/score/metascore/all/filtered?sort=desc&page=' + str(i)
    all_urls.append(url)
print(all_urls)
```

```python
#we need to use a user agent here, because metacritics does not accept bots
#we are using a tested go-to user agent here, it can be found online

headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36'}
```

```python
# Iterating over 5 page links
# save all the information from the url into a container

movie_names = []
movie_links = []
for i in all_urls:
    html = requests.get(i,headers = headers)
    soup = BeautifulSoup(html.content, "html.parser")
    container = soup.find_all('td',class_ = "clamp-summary-wrap")
    
# Saving All movie names in a list
    for movie in container:
        name = movie.find('h3').text
        movie_names.append(name)
        
#Extracting the movie URL and saving in a list
        tag = movie.find('a', class_ = 'title')
        link = tag.get('href', None)
        movie_links.append(link)
        
print(len(movie_names))
print(len(movie_links))   
```

```python
# We can see the link is not complete
# Add https://www.metacritic.com to each element in list movie_links

for i in range(len(movie_links)):
    movie_links[i] = 'https://www.metacritic.com' + movie_links[i]
print(movie_links)
```

```python
# We need to retain the details of each movie form detail links 
# each movie has a seperate detail link

detail_links = []
for movie_link in movie_links:
    page = requests.get(movie_link, headers = headers)
    soup = BeautifulSoup(page.content, "html.parser")
    div = soup('div', class_= 'see_all pad_btm1')
    for line in div:
        hyperlink = line.find('a')
        url = hyperlink.get('href', None)
        detail_links.append(url)
        print(url)
print(len(detail_links))
```

```python
#completing the detail links by adding https://www.metacritic.com to each link

for i in range(len(detail_links)):
    detail_links[i] = 'https://www.metacritic.com'+ detail_links[i]
print(len(detail_links))
```

```python
# After retrieving the urls of detail pages of each movie
# We need to extract the html contents of the page

directors = list()
director_m = list()

for url in detail_links:
    page = requests.get(url,headers = headers)
    soup = BeautifulSoup(page.content,'html.parser')
    
# we need to extract the director info
# but we notice that there are 5 tables, all of them named 'credits'
# so we need one more information, the summary
# we use re.compile ro compile all the tables with director's credits

    director_table = soup.find_all('table', attrs = {'summary': re.compile('Director Credits')})
    
#now that we have director table
#we can add a step to extract people from it
#checking if the movie has more than one director
#for the movies with more than one director we append them as list of directors in the list

    if len(director_table) != 0:
        for direc in director_table:
            director_line = direc.find_all('a')
            if len(director_line) == 1:
                for director in director_line:
                    directors.append(director.text.strip())
            else:  
                for director in director_line:
                    director_m.append(director.text.strip())
                directors.append(director_m)
                director_m = []
```

```python
# list of all the directors
print(directors)
print(type(directors))
print(len(directors))

# importing groupby from itertools to take out the redundency casues by similar movie names

#from itertools import groupby
#director = [i[0] for i in groupby(directors)]
#print(directors)
#print(len(directors))

# saving the list of directors as tuple
ed = []
for i in directors:   
    if type(i) == list:
        i = tuple(i)
    ed.append(i)
ed = tuple(ed)             
print(ed)

flat_directors = []
for i in directors:
    if type(i) == list:
        for j in i:
            flat_directors.append(j)
    else:
        flat_directors.append(i)
        
print (flat_directors)
```

```python
# extracting the html contents of the page to retrieve cast information

casts = list()
full_cast = list()

for url in detail_links:
    page = requests.get(url,headers = headers)
    soup = BeautifulSoup(page.content,'html.parser')
# extracting the cast info using the summary "cast credits tag"
# some of the cast_table contents are None type
# some movies do not have any cast info in metacritics
# We create an empty list if there is no cast info

    cast_table = soup.find_all('table', attrs = {'summary': re.compile('Cast Credits')})

    if len(cast_table) != 0:
        if len(cast_table) != 0:
            for cast in cast_table:
                
                # extract all the contents of tag a
                #principle cast and cast are present in tag a
                # then strip it and save it in an empty list
                
                cast_line = cast.find_all('a')  
                for cast in cast_line:
                    casts.append(cast.text.strip())
            #append the each cast of movie to main list
            #empty the temporary list for each iteration
            full_cast.append(casts)
            casts = []
    else:
        full_cast.append([])
     
```

```python
# extracting the html contents of the page to retrieve cast information

credits = list()
full_credits = list()

# extracting the credits info using the summary "cast credits tag"
# some of the credit_table contents are None type
# some movies do not have any cast info in metacritics
# We create an empty list if there is no credit info

for url in detail_links:
    page = requests.get(url,headers = headers)
    soup = BeautifulSoup(page.content,'html.parser')
    
    credit_table = soup.find_all('table', attrs = {'summary': re.compile('Cast Credits')})

    if len(credit_table) != 0:
        if len(credit_table) != 0:
            for credit in credit_table:
                # extract all the contents of tag <'td', class_ = 'role'>
                # principle cast and cast credits are present in tag <'td', class_ = 'role'>
                # credits do not have a tag a like cast info
                # then strip it and save it in an empty list
                
                credit_line = credit.find_all('td', class_ = 'role')  
                for credit in credit_line:
                    credits.append(credit.text.strip())
                    
            #append the each cast of movie to main list
            #empty the temporary list for each iteration
            
            full_credits.append(credits)
            credits = []
    else:
        full_credits.append([])
     
```

```python
#then we need to create the combined list containing all actors that the 2 directors have worked with
#first we extract the actor names, which are the keys
sub_actor1 = list(dir1_count.keys())
sub_actor2 = list(dir2_count.keys())

#then we convert them into sets, so that we can find a combined set using the union function
set_actor1 = set(sub_actor1)
set_actor2 = set(sub_actor2)
combine_set = set_actor1.union(set_actor2)

#we make two directories with actors/actresses from combined set as keys and 0 as values
dir1_actor = dict()
for i in combine_set:
    dir1_actor[i] = 0

dir2_actor = dict()
for i in combine_set:
    dir2_actor[i] = 0


# now we append original values from count to these dictionaries
for key, value in dir1_actor.items():
    for k, v in dir1_count.items():
        if key == k:
            dir1_actor[key] = v


for key, value in dir2_actor.items():
    for k, v in dir2_count.items():
        if key == k:
            dir2_actor[key] = v

#now that we have a uniform dictionary across the 2 directors, we can extract the values in vectors
#ready for comparison
dir1_vector = list(dir1_actor.values())
dir2_vector = list(dir2_actor.values())

#calculate cosine similarity
divident = 0 
for i in range(len(dir1_vector)):
    summing = dir1_vector[i] * dir2_vector[i]
    divident = divident + summing


a = 0
b = 0
for i in range(len(dir1_vector)):
    sum1 = (dir1_vector[i])**2
    a = a + sum1
    sum2 = (dir2_vector[i])**2
    b = b + sum2
divisor = (a**0.5)*(b**0.5)


cosine_similarity = divident/divisor
print("Similarity score of the choosen directors", director1, "and", director2, "is :",  cosine_similarity)
```
